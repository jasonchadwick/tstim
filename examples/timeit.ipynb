```python
import timeit
import numpy as np
import random
```

```python
timeit.timeit(lambda: set([1, 2, 3, 3, 2, 3]), number=100000)/100000
```

```python
timeit.timeit(lambda: np.unique(np.array([1, 2, 3, 3, 2, 3])), number=100000)/100000
```

```python
timeit.timeit(lambda: list(set(np.array([1, 2, 3, 3, 2, 3]))), number=100000)/100000
```

```python
timeit.timeit(lambda: len(set([False, False, False, True])), number=100000)/100000
```

```python
timeit.timeit(lambda: np.all([False, False, False, True]), number=100000)/100000
```

```python
timeit.timeit(lambda: len([x for x in [False, False, False, True] if x]), number=100000)/100000
```

```python
timeit.timeit(lambda: np.random.choice(255, 250), number=100000)/100000
```

```python
timeit.timeit(lambda: random.choices(range(255), k=250), number=100000)/100000
```
